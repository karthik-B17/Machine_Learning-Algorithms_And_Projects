# Pareto Distribution

Also known as power law distribution 
It is a continuous distribution that models the phenomena where the small proportion of values account for large population of effect

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto
```

```python
# parameter
alpha=2.62
x=np.linspace(1,10,1000)
pdf=pareto.pdf(x,alpha)
cdf=pareto.cdf(x,alpha)
pdf,cdf
```

```python
plt.plot(x,pdf,label='PDF',color='blue')
plt.title('Pareto Distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid()
plt.legend()
```

```python
plt.plot(x,cdf,label='CDF',color='orange')
plt.title('Pareto Distribution')
plt.xlabel('x')
plt.ylabel('Density') 
plt.grid()
plt.legend()
```

```python
import seaborn as sns
```

```python
from scipy import stats
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
tot_bill_th=stats.norm.rvs(loc=tips["total_bill"].mean(),scale=tips["total_bill"].std(),size=len(tips))
```

```python
tot_bill_th
```

```python
# plot histogram,kde,and boxplot for the theoretical total bill
fig,ax=plt.subplots(1,3,figsize=(16,7))
sns.histplot(tot_bill_th,ax=ax[0],kde=True)
sns.kdeplot(tot_bill_th,ax=ax[1],fill=True)
sns.boxplot(tot_bill_th,ax=ax[2])
plt.show()
```

```python
# plot histogram,kde,and boxplot for the theoretical total bill
fig,ax=plt.subplots(1,3,figsize=(16,7))
sns.histplot(tips["total_bill"],ax=ax[0],kde=True)
sns.kdeplot(tips["total_bill"],ax=ax[1],fill=True)
sns.boxplot(tips["total_bill"],ax=ax[2])
plt.show()
```

```python
# Verifying the normality using QQ plot
import statsmodels.api as sm

sm.qqplot(tips["total_bill"],line='s')
plt.title('QQ plot for total bill')
```

```python
stats.probplot(tips['total_bill'],dist=stats.norm,plot=plt)
plt.title('Probability plot for total bill')
plt.show()
```

```python
stats.probplot(tot_bill_th,dist=stats.norm,plot=plt)
plt.title('Probability plot for total bill')
plt.show()
```

```python
sm.qqplot(tot_bill_th,line='s')
plt.title('QQ plot for theoretical total bill')
plt.show()
```

How to compare:

1.Quantile Calculation: Both the observed data and the theoretical normal distribution are divided into equal-sized portions (quantiles).

2.Plotting: The quantiles of the observed data are plotted on one axis, and the corresponding quantiles of the theoretical normal distribution
are plotted on the other axis. 

3.Interpretation: If the two distributions are similar, the plotted points will form a straight line. 

4.Deviation from the line: indicates that the observed data's distribution differs from the normal distribution. 

5.Patterns: in the deviation, such as curves or an "S" shape, can reveal the specific nature of the non-normality.
