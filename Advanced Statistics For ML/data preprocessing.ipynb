# Data Preperation:

1. Identifying the X(input features) and Y(Target feature).
2. Perform train test(val) split.
3. Preprocess training data.
4. Preprocess test data.|

```python
x=df.iloc[:,:4]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
# data transformation

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
#sc.fit(x_train)
x_train_data=sc.fit_transform(x_train)
x_test_data=sc.transform(x_test)
```
