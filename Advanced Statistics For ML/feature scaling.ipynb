# Standardization:

- Its a scaling technique where the values are centered around the mean and the unit standrard deviation.
- The mean of the feature becomes zero and the resultant distribution has a unit standard deviation.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("C:\\Users\\karth\\Downloads\\train.csv",usecols=['Age','Fare']  )
df
```

```python
from sklearn.preprocessing import StandardScaler 
sc=StandardScaler()
```

```python
df_new=pd.DataFrame(sc.fit_transform(df),columns=df.columns)
df_new
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 4))
ax1.set_title('Before Standardization')
sns.kdeplot(data=df, x="Age", ax=ax1)
sns.kdeplot(data=df, x="Fare", ax=ax1)  
ax2.set_title('After Standardization')
sns.kdeplot(data=df_new, x="Age", ax=ax2) 
sns.kdeplot(data=df_new, x="Fare", ax=ax2)  
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14,7))

plt.subplot(1,2,1)
plt.title("Scatterplot before standardization")
sns.scatterplot(data=df,color="blue")


plt.subplot(1,2,2)
plt.title("Scatterplot after standardization")
sns.scatterplot(data=df_new,color="red")
```


- mean pf distribution is very close to 0 and std is exactly 1.
- some outliers in Fare and that's why the mean didnt reduce to exactly 0.
- Scatterplot- the scale changed and the distribution came to centre or 0.
- kde- exactly same so this shows that distribution is not effected by standardization.


# Normalization:

- The goal of normalizationis to change the values of numeric columns in the dataset to use a common scale, without disturbing
  differences in the range of values or losing information.

  ## 1. Min-max scaling:

- min value will transformed into 0
- max value will transformed into 1
- any other value will get transformed into decimal ranging between 0-1.

x'=x-min(x)/max(x)-min(x)

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
df_new_mm=pd.DataFrame(mm.fit_transform(df),columns=df.columns)
```

```python
df_new_mm
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(14, 4))
ax1.set_title('Before MinMax Scaling')
sns.kdeplot(data=df, x="Age", ax=ax1)
sns.kdeplot(data=df, x="Fare", ax=ax1)  
ax2.set_title('After MinMax Scaling')
sns.kdeplot(data=df_new_mm, x="Age", ax=ax2) 
sns.kdeplot(data=df_new_mm, x="Fare", ax=ax2)  
plt.show()
```

```python
plt.figure(figsize=(14,4))

plt.subplot(1,2,1)
plt.title("Scatterplot before MinMax scaling")
sns.scatterplot(data=df,color="blue")


plt.subplot(1,2,2)
plt.title("Scatterplot after MinMax scaling")
sns.scatterplot(data=df_new_mm,color="red")
```

- Min-max normalization works when the min and max is very distinct and known
