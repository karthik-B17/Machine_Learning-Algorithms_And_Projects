# Exponential Distribution:

- Exponential is single parameter continuous distribution that is traditionally used for modeling time to failure of electronic
 components.

- it represents a process in which event occurs continuously and independently at a constant average rate.

```python
parameter 1: parameter (lambda) is the scale parameter and represents the rate of occurrence of the event.
parameter 2: Mean of exponential distribution is given by 1/lambda
```

```python

The time-to-failure of an avionic system follows an exponential distribution with a mean time between
failures (MTBF) of 1000 hours. Calculate
1. The probability that the system will fail before 1000 hours.
2.The probability that it will not fail up to 2000 hours.
3.The time by which 10% of the system will fail (i.e., calculate P10 life).
```

```python
#The probability that the system will fail before 1000 hours.
from scipy import stats
stats.expon.cdf(1000,loc=1/1000,scale=1000)
```

```python
# The probability that it will not fail up to 2000 hours.

#finding thd probability that it will fail up to 2000 hours

stats.expon.cdf(2000,loc=1/1000,scale=1000)
```

```python
1-stats.expon.cdf(2000,loc=1/1000,scale=1000)
```

```python
# .The time by which 10% of the system will fail (i.e., calculate P10 life).

stats.expon.ppf(0.1,loc=1/1000,scale=1000)

#by 105.36 hours 10% of the system will fail.
```

```python
# plotting the graph for exponential:
import pandas as pd
pdf_df=pd.DataFrame({"Success":range(0,5000,100),"pdf":stats.expon.pdf(range(0,5000,100),loc=1/1000,scale=1000)})
pdf_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(22,5))
sns.barplot(x="Success",y="pdf",data=pdf_df)
```
