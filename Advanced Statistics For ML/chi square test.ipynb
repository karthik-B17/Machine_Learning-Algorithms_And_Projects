Chi-square test of independence

1. purpose: Test the association between two categorical variables.
2. When to use: Use for categorical variables in a contigency table.
3. Assumptions: Data is in contigency table format,Expected fequencies in the contigency table >5.

```python
import seaborn as sns
import pandas as pd
from scipy.stats import chi2_contingency

df=sns.load_dataset("tips")
df=pd.DataFrame(df)
```

```python
df.head(10)
```

```python
# claim

#Gender and smoking habits in tips dataset are independent
```

Â· H0: Gender and smoking habits are independent (This means that distri of amoking habits doesnot differ between genders)
. H1:Gender and smoking habits are not independent( This means that smoking habits depend on gender (i.e., there's an association between gender and smoking
habits).

```python
import pandas as pd
from scipy import stats
```

```python
contingency_table=pd.crosstab(df["sex"],df["smoker"])
contingency_table
```

Given significance level is 0.05

```python
from scipy.stats import chi2_contingency
```

```python
chi2_value,p_value,df,exp_values=chi2_contingency(contingency_table)
chi2_value,p_value
```

# chi-square goodness of fit

Claim:

The proportion of male and female in the tips dataset in 50-50

. H0:The observed porportion of male and female match the expected proportion of(50-50).
. H1:The observed porportion of male and female does not match with the expected proportion of(50-50).




```python
df=sns.load_dataset("tips")
sample=df["sex"]
observed=sample.value_counts()
observed
```

```python

expected=[len(sample)/2]
expected
```

```python
chi2_stat,p_value=stats.chisquare(f_obs=observed,f_exp=expected)
chi2_stat,p_value
```
