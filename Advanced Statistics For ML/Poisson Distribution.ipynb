```python
Poisson distribution is discrete probability distribution in the situtations where we are intrested in calculating the number of events 
that may occur over a period of time or space.

for example:
number of cancellations of order by customer in an hour
or number of call received in a call centre.
```

```python

The number of calls arriving at a call center follows a Poisson distribution at 10 calls per hour.
1.Calculate the probability that the number of calls will be maximum 5.
2.Calculate the probability that the number of calls over a 3-hour period will exceed 30.
```

```python
1. first parameter: Number of events for which probability needs to get calculated.(k) --->5
2. second parameter: The average number of events.(lambda)------>10
```

```python
# 1.Calculate the probability that the number of calls will be maximum 5.

from scipy import stats
stats.poisson.cdf(5,10)
```

```python
# the chances of getting maximum of 5 calls is near about 6%.
```

```python
#probability of maximum 30 calls in 3 hours which makes the average number of calls for 3 hours as 30. (lambda)
stats.poisson.cdf(30,30)
```

```python
#probability of more than 30 calls in 3 hours 
1-stats.poisson.cdf(30,30)
```

```python
# plotting the poisson distribution for the given data.

#no of calls (0,40)--->x axis
# y axis---->pmf
import pandas as pd

pmf_df=pd.DataFrame({"Success":range(0,41),"pmf":stats.poisson.pmf(range(0,41),10)})
pmf_df
```

```python
#creating a bar plot with no. of calls in x axis and pmfs in y
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sns.barplot(x="Success",y="pmf",data=pmf_df)
```

```python
pmf_df_3=pd.DataFrame({"Success":range(0,41),"pmf":stats.poisson.pmf(range(0,41),30)})
pmf_df_3
```

```python
#creating a bar plot with no. of calls in x axis and pmfs in y
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sns.barplot(x="Success",y="pmf",data=pmf_df_3)
```

```python
A bakery receives on average 2 online cake orders per day. Assume the number of cake orders follows a Poisson distribution.
Question 1: What is the probability that the bakery gets no orders in a day?
Question 2: What is the probability that the bakery gets at least 3 orders in a day?
```

```python
# probability of getting exactly 0 orders in a day.
stats.poisson.pmf(0,2)  #13.53%
```

```python
# probability that we are getting maximum of 2 orders in a day
stats.poisson.cdf(2,2)
```

```python
# probability that the bakery gets at least 3 orders in a day
1-stats.poisson.cdf(2,2)
```
