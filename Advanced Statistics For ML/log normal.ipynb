

Log Normal:

Â· The lognormal distribution is a continuous probability distibution of random variable whos logarithm is normally distributed.
. In simple terms Y=log(X) is normally distributed, then X follows a lognormal distribution.

Plotting PDF and CDF for log normal distribution:

```python


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

```

```python
# Parameters for lognormal
# mean of underlying normal distribution
mean_sigma=0.5 # standard dev of underlying normal distribution

mean_mu=0
```

```python
x=np.linspace(0.01,5,1000)
x
```

```python
pdf=lognorm.pdf(x,mean_sigma,scale=np.exp(mean_mu))
cdf=lognorm.cdf(x,mean_sigma,scale=np.exp(mean_mu))
pdf,cdf
```

```python
plt.plot(x,pdf,label='PDF',color='blue')
```

```python
plt.plot(x,cdf,label='CDF',color='blue')
```

An investor is analyzing the future value of a stock investment. It is known from historical data that the annual return multiplier
(i.e., the ratio of value at year-end to start-of-year) follows a log-normal distribution.
For a particular stock:
The natural logarithm of the return multiplier follows a normal distribution with:
Mean u=0.07

Standard deviation o=0.2

This implies the return multiplier (lets call it R) follows a LogNormal(u=0.07, o=0.2).
Questions:
1.What is the probability that the stock returns less than 5% in a year (i.e., R<1.05)

2.What is the median return multiplier? 

3. Plot the PDF and CDF of the return multiplier.

```python

# since the return multiplier follows a log normal distribution so 5% is also log-nomal value which in the question becomes 1.05 |
# nomral distribution.

mean=0.07
STD=0.2
threshold=1.05

lognorm.cdf(1.05,s=STD,scale=np.exp(mean))
```

```python
lognorm(s=STD,scale=np.exp(mean)).median()
```

```python
x=np.linspace(0.01,5,1000)
x
```

```python
pdf=lognorm.pdf(x,STD,scale=np.exp(mean))
cdf=lognorm.cdf(x,STD,scale=np.exp(mean))
pdf,cdf
```

```python
plt.plot(x,pdf,label='PDF',color='blue')
plt.plot(x,cdf,label='CDF',color='orange')
```
