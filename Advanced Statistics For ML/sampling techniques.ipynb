# Sampling Techniques:

. Sampling techniques from the foundation of statiatical analysis and research methodologies.
. These techniques provides the systematic ways to select a subset(sample) from a larger group (population) for analysis.
. By working with a sample researchers can infer characteristics about the entire population without the need to study every individual or item, which may be
impractical due to constraint like time,cost and feasibility.
## 1. Simple Random Sampling:
- ** Description ** : selects items randomly, ensuring each has an equal chance of being chosen.
- ** Methods used ** :
1. np.choose: selects random indices from sampling.
2. dataframe.sample: randomly picking rows form the dataset.
- ** Strengths ** : Highly unbaised and representative.
- ** Limitations ** : Doesnt gurantee subgroups representation and require complete list of population.|

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
df=titanic[['sex','age','fare']]
```

```python
df
```

```python
n_sample=10

random_indices=np.random.choice(df.index,size=n_sample,replace=False)
random_indices
```

```python
sample_np=df.loc[random_indices]
sample_np
```

```python
simple_random_df=df.sample(n=n_sample,random_state=33)
simple_random_df
```

Systematic Sampling:

. Introduces the structure by selecting items at fixed intervals.
. May fail if the data has periodic patterns.



```python
# suffle and slice in the steps of n

n_step=20

shuffled_df=df.sample(frac=1,random_state=80).reset_index(drop=True)
```

```python
shuffled_df
```

```python
systematic_slicing=shuffled_df[::n_step]
systematic_slicing
```


Stratified Sampling.

· Gurantees subgroup representation (e.g. equal sampling from Male and female groups)
. More complex to implement but useful when population has distinct subgroups.

# group by strata (sex) and perform simple random sampling

```python


strata="sex"

stratified_sample=df.groupby(strata,group_keys=False).apply(lambda x:x.sample(n=5,random_state=42))
```

```python
stratified_sample
```

Clustered Sampling:

- Population is divided into non-overplapping clusters.
- Some clusters are randomly selected
- Either all individuals in seleted cluster are surveyed (one-stage) or sample is take from each selected cluster (two-stage)

```python
titanic
```

```python
clusters=titanic['class'].unique()
```

```python
 # step2: randomly select 1 cluster (day)
np.random.seed(43)
select_cluster=np. random.choice(clusters,size=1)
select_cluster

# select all data from that cluster.

cluster_sample=titanic[titanic["class"].isin(select_cluster)] 
cluster_sample
```
