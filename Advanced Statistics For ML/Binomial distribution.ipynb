Binomial distribution is a discrete probability distribution and has several applications in many business contexts.
A random variable X is said to follow a binomial distribution when:

1. The Random variable can have only two outcomes-- success and failure.( also known as bernoulli trials)
2. The objective is to find the probability of getting x successes out of n trials.
3. The probability of success is "p" and thus the probability of failure is (1-p).
4. The probability "p" is constant and does not change between trials.

```python
examples:

1. Customer churn where the outcomes are a. Customer churn b. No customer churn
2. Loan repayment default by a customer a. Default b. Not default
```

## 1.Probability that exactly 5 customers will return the items.

```python
1.expected number of successful trials----> 5
2. Total number of trials------->20
3. probability of success----->0.1
```

```python
from scipy import stats
stats.binom.pmf(5,20,0.1)
```

```python
The corresponding probability is 0.03192 that is the probability that exactly 5 customers will return the items is approximately 3%.
```

## 2.Probability that a maximum of 5 customers will return the items.

```python
# range(0,21)
import pandas as pd
pmf_df=pd.DataFrame({"Success":range(0,21),"PMF":stats.binom.pmf(range(0,21),20,0.1)})
pmf_df
```

```python
# creating a bar plot
import seaborn as sns
sns.barplot(x="Success",y="PMF",data=pmf_df)
```

```python
# in the question it says probability of success is 10% which means the heighest probabilty of customers out of 20 customers who will 
#return thr item is 2 as shown in the above graph.

stats.binom.cdf(5,20,0.1)
```

## 3.Probability that more than 5 customers will return the items purchased by them.

```python
1-stats.binom.cdf(5,20,0.1)
```

## 4.Average number of customers who are likely to return the items and the variance and the standard deviation of the number of returns.

```python
Average of binomial distribution is given by = n*p
variance of binomial distribution is given by = n*p*(1-p)
```

```python
mean,var=stats.binom.stats(20,0.1)
mean
```

```python
var
```

```python
import math
math.sqrt(var)  # ----->std
```
