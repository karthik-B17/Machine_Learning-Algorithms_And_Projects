# Hypothesis testing:

- Hypothesis testing is a core statistical method used to make iferences about a population based on the sample.
- It evaluates the claims, test relationships and determine whether observed results are statistically significant.

## Z test is used when:

1. We need to test the value of population mean given that population variance/std is known.
2. The population follows the Normal distribution and variance is known.
3. Sample size is large n>30.

A passport office claims that the passport applications are processed within 30 days of
submitting the application form and all necessary documents. The file passport.csv contains
processing time of 40 passport applicants. The population standard deviation of the
processing time is 12.5 days. Conduct a hypothesis test at significance level a = 0.05 to verify
the claim made by the passport office.
In this case, the population mean (claim made by passport office) and standard devia-
tion are known. Population mean is 30 and population standard deviation is 12.5. The data-
set in passport.csv contains observations of actual processing time of 40 passports. We can
calculate the mean of these observations and calculate Z-statistics.

population std=12.5
Population mean =30
H0: mean>30
Ha:mean <= 30

Trick:
In Z-test
What ever in the question that have been given is considered as alternative hypothesis and the opposite is considered as null hypothesis and we prove that if the probability is less than the significane level then we say that the null hypothesis is wrong else true

```python
pop_std=12.5
pop_mean=30
# one tailed left side Scenario(Hypothesis)
```

```python
import math
from scipy import stats
def ztest(pop_mean,pop_std,sample):
    z_score=(sample.mean()-pop_mean)/(pop_std/math.sqrt(len(sample)))
    return z_score,stats.norm.cdf(z_score)
```

```python
ztest(30,12.5,df["Processing_time"])
```

A company claims that their energy drink increases alertness for at least 3 hours on average.
A consumer advocacy group believes it's actually less than 3 hours and wants to test the claim.
They test a random sample of 40 people, and find that the average time of increased alertness is 2.7 hours,
with a known population standard deviation of 0.8 hours.
At a 5% significance level, test the claim using a one-tailed Z-test.

population_std=0.8
n=40
sample_mean=2.7
SL=5%
CI=95%

```python
import math
from scipy import stats
def ztest(sample_mean,pop_mean,pop_std,sample):
    z_score=(sample_mean-pop_mean)/(pop_std/math.sqrt(sample))
    return z_score,stats.norm.cdf(z_score)

ztest(2.7,3,0.8,40)
```
