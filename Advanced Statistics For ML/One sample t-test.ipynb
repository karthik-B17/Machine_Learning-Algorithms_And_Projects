# one sample t-test:

purpose: whether the mean of single sample is significantly different from a known or hypothesized population mean.
When to use: continuous data when comparing a sample mean to a specific value.

Assumptions:
1. Data follows a normal distribution.
2. Observations are independent.|

- Example: The average total bill in the tips dataset is not 20 rupees.

- H0: he average total bill in the tips dataset is 20 rupees.
- H1: The average total bill in the tips dataset is not 20 rupees.

how to decide hypothesis.

HÃ˜: there is no effect or difference or the value is as claimed.
H1: it reflects the claim which you are trying to test.

mean=20
mean != 20

```python
from scipy import stats
```

```python
import seaborn as sns
df=sns.load_dataset("tips")
df.head(10)
```

```python
# The average total bill in the tips dataset is 20 rupees
t_value,p_value=stats.ttest_1samp(df['total_bill'],20)
print(t_value,p_value)
```

# Two sample t-test:

Purpose: compares the mean of two related samples.
When to use : use for categorical and numerical data

Assumptions:
1.Both sample should be normally distributed.
2. variance of two samples must be equal.

syntax: stats.ttest_ind(sampl_1,sample_2)

Claim :The average total bill differs between smokers and non smokers.

. H0:The average total bill does not differ between smokers and non smokers.
. H1:The average total bill differs between smokers and non smokers.

```python
smokers=df[df["smoker"]=="Yes"]["total_bill"]
smokers
```

```python
non_smokers=df[df["smoker"]=="No"]["total_bill"]
non_smokers
```

```python
t_value,p_value=stats.ttest_ind(smokers,non_smokers)
print(t_value,p_value)
```

# as we can see the p_value is greater than that of significance value which means we are going to accept the null hypothesis stating
# that average total_bill of smokers and non smokers doesnot differ that much.
