```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(df,x="f1", y="y")
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features=PolynomialFeatures(degree=2)
x_poly=poly_features.fit_transform(df [["f1"]])
x_poly
```

```python

# transforming the above array into df
transformed_df=pd.DataFrame(x_poly,columns=["f1_0","f1_1","f2_2"])
transformed_df ["y"]=df["y"]
transformed_df
```

```python
sns.pairplot(transformed_df)
```

```python
reg_poly=LinearRegression()
reg_poly.fit(transformed_df[["f1_0","f1_1","f2_2"]],transformed_df["y"])
```

```python
# pred
pred_poly=reg_poly.predict(transformed_df[["f1_0","f1_1","f2_2"]])
```

# Other Regression Models(Lasso,Ridge And Elastic Net)

```python
from sklearn.linear_model import Lasso,Ridge,ElasticNet # Lasso L1-Sparsity in data to reduce  the dimensionality
```

```python
reg_lasso=Lasso(alpha=1.0)
reg_lasso.fit(transformed_df[["f1_0","f1_1","f2_2"]],transformed_df["y"])
```

```python
pred_lasso=reg_lasso.predict(transformed_df[["f1_0","f1_1","f2_2"]])
```

```python
pred_lasso
```

```python
plt.scatter(df["f1"],df["y"])

plt.plot(df["f1"],pred_lasso,"green")



```

```python
# ridge implementation

reg_ridge=Ridge(alpha=1.0)

reg_ridge.fit(transformed_df[["f1_0","f1_1","f2_2"]],transformed_df["y"])
```

```python
pred_ridge=reg_ridge.predict(transformed_df[["f1_0","f1_1","f2_2"]])
```

```python
plt.scatter(df["f1"],df["y"])

plt.plot(df["f1"],pred_ridge,"brown")
```

```python
# ElasticNet implementation:

reg_elastic=ElasticNet(alpha=1.0)

reg_elastic.fit(transformed_df[["f1_0","f1_1","f2_2"]],transformed_df["y"])
```

```python
pred_elastic=reg_elastic.predict(transformed_df[["f1_0","f1_1","f2_2"]])
```

```python
plt.scatter(df["f1"],df["y"])

plt.plot(df["f1"],pred_elastic,"red")
```
