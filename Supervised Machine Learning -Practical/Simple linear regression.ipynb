```python
x=df.iloc[:,0]
y=df.iloc[:,1]
x,y
```

```python
# for linear regression we need to have the dependency among the x input feature and y target column
import seaborn as sns
sns.scatterplot(data=df,x="Hours",y="Scores")
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.7,random_state=56)
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

```python
x_train=pd.DataFrame(x_train)
type(x_train)
```

```python
y_train=pd.DataFrame(y_train)
type(y_train)
```

```python
regressor.fit(x_train,y_train)
```

```python
# intercept of the best fit line
regressor.intercept_
```

```python
x_test=pd.DataFrame(x_test)
```

```python
#making the predictions of the model
y_pred=regressor.predict(x_test)
y_pred
```

```python
y_test
```

```python
regressor.score(x_test,y_pred)
```

```python
#residual analysis of the train data:
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train)
plt.plot(x_train,regressor.predict(x_train),color='red')
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred,color='blue')
```

```python
# evaluation of the algorithm
from sklearn import metrics
metrics.mean_absolute_error(y_test,y_pred)
```

```python
metrics.mean_squared_error(y_test,y_pred)
```

```python
import numpy as np
np.sqrt(metrics.mean_squared_error(y_test,y_pred))
```
